```python
import json
import requests
import pandas as pd
```

```python
dates_list = pd.date_range(start="03/03/2019", end="03/16/2021")

dates = []
for d in dates_list:
    date = str(d)
    date = date[:10]
    date = date.replace('-', ' ')
    parts = date.split(sep=' ')
    dates.append(str(parts[1])+ str(parts[2])+ str(parts[0]))
```

```python
len(dates_list)
```

```python
file_num=1
no_data_dates=[]


for d in dates:
    try:
        #url = "https://stocknewsapi.com/api/v1?tickers=FB&items=50&token=yd7wjd23l89iwqcc7vvm41llegsmuefsxw5322zp&date={0}-{0}&datatype=csv".format(d)
        url = "https://stocknewsapi.com/api/v1/category?section=general&items=50&token=yd7wjd23l89iwqcc7vvm41llegsmuefsxw5322zp&date={0}-{0}&datatype=csv".format(d)
        df = pd.read_csv(url) 
        df.to_csv("./data/general-news/{0}.csv".format(file_num), index=False)
        file_num+=1
        print(file_num)
    except:
        no_data_dates.append(d)
        print("nothing on ", d)
        continue
```

```python
import os
import glob

path = r'./data/general-news/'                     # use your path
all_files = glob.glob(os.path.join(path, "*.csv"))     # advisable to use os.path.join as this makes concatenation OS independent
```

```python
all_files
```

```python
df_from_each_file = (pd.read_csv(f) for f in all_files)
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)
```
